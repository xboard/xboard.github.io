   >> Considere um supermercado de classe média (ou um colégio particular, ou um clube, ou um restaurante). Assuma como *prior*, que o pais é **justo** e que há iguais chances de você ver brancos e negros entre os clientes  do supermercado e que a *população afrodescendente é de ao menos 50% na nossa população*. Agora, tome uma observação: suponha que você observou 200 clientes no supermercado e dentre eles apenas 10 eram afrodescendentes. Como isso afeta seu posterior, isto é, como você revê sua crença na igualdade considerando tais dados?

```python
%matplotlib inline
from IPython.core.pylabtools import figsize
from IPython.html.widgets import interact
import numpy as np
from matplotlib import pyplot as plt
figsize(11, 9)

import scipy.stats as stats

dist = stats.beta
x = np.linspace(0, 1, 100)

def plot_posterior(**kwargs):
    populacao = kwargs['populacao']
    num_ad = kwargs['afrodescendentes']
    
    fig, ax = plt.subplots(figsize=(10, 5),
                           subplot_kw={'axisbg':'#EEEEEE',
                                       'axisbelow':True})
    ax.grid(color='w', linewidth=2, linestyle='solid')
    
    y = dist.pdf(x, 1 + num_ad, 1 + populacao - num_ad)
    ax.plot(x, y, label="Universo de %d pessoas,\n %d afrodescendentes vistos" % (populacao, num_ad))
    ax.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    ax.vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)

    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    ax.autoscale(tight=True)

i = interact(plot_posterior, populacao=(1,400), afrodescendentes=(1,200))
```

# Comentários

Destrinchando a ótima solução do [Flávio](https://github.com/xboard). Considere $\theta$ a proporção de afrodescendentes na população e, consequentemente, o a proporção de clientes que esperaríamos ver no estabelecimento. Seja $y$ o número de afrodescendentes observados em $n$ clientes. A verossimilhança é dada por uma distribuição binomial:

$p(y|\theta) = \text{Bin}(y|n, \theta) = \binom{n}{y} \theta^y (1 - \theta)^{n-y}$

Assumindo um *prior* uniforme no intervalo $[0,1]$, o Flávio matou o problema: o *posterior* é proporcional a $\text{Beta}(y+1, n - y + 1)$ (ver Gelman *et al.*, Seção 2.1 para detalhes).

```python
y = 10
n = 200
posterior = stats.beta(y+1, n-y+1)
```

```python
theta = linspace(0., 1., 100)
figsize(10,5)
plot(theta, posterior.pdf(theta))
xlabel(r'$\theta$') 
ylabel(r'$\propto p(\theta|y)$')
```

O primeiro gráfico, criado pelo Flávio, é interativo, permitindo brincarmos com $y$ e $n$, que é o espírito deste exercício. O *prior* utilizado foi uma distribuição uniforme no intervalo $[0,1]$, ou seja, assumimos que não temos qualquer crença inicial no valor de $\theta$.

O gráfico acima nos diz que é *implausível* que a proporção de afrodescentes entre os clientes do estabelecimento seja superior a 10%. De acordo com o IBGE, 51% da população se declara não-branca, de modo que teríamos, **de acordo com esses valores de $y$ e $n$**, um cenário claro de segregação. 

## Um *prior* diferente

E se assumirmos um *prior* em torno de 51%?

```python
nonwhite_prior = stats.norm(0.51, 0.05)
```

```python
plot(theta, nonwhite_prior.pdf(theta))
xlabel(r'$\theta$')
ylabel(r'$\mathrm{pdf}(\theta)$')
```

O gráfico acima nos diz que temos bastante confiança, a princípio, que a proporção de não-brancos no estabelecimento esteja entre 40% e 60%. A beleza da análise Bayesiana é que ela permite que os dados compensem crenças iniciais distorcidas:

```python
def nonwhite_posterior_propto(theta):
    nonwhite_likelihood = stats.binom(n, theta)
    return nonwhite_prior.pdf(theta) * nonwhite_likelihood.pmf(y)
```

```python
plot(theta, nonwhite_posterior_propto(theta))
xlabel(r'$\theta$')
ylabel(r'$\mathrm{pdf}(\theta | y)$')
```

Mesmo que você acredite **muito** que há igualdade de oportunidades para estar no referido estabelecimento, os dados urgem para que você mude de ideia (Keynes mudaria!). O gráfico acima mostra que dificilmente haveria uma probabilidade superior a 20% de encontrarmos não-brancos no tal estabelecimento, uma segregação clara em um país onde a população não-branca é superior a 50%.
